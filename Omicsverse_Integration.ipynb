```python
bulk_data=ov.read('C:Desktop/GerardoIga/2024/Drosophila/RNA-seq/BrainWT-0224-2024/Count_table_geneid_Fly_brain_M_F_Capsaicin_RNASeq.txt',index_col=0)
bulk_data.head()
```

```python
adata_raw = adata.raw.to_adata()
```

```python
adata_new=ov.pp.preprocess(adata_raw,
                       n_HVGs=2000,batch_key=None)
adata_new
```

```python
%%time
adata_new.raw = adata_new
adata_new = adata_new[:, adata_new.var.highly_variable_features]
adata_new
```

```python
model=ov.bulk2single.Bulk2Single(bulk_data=bulk_data,single_data=adata_new, 
                celltype_key='Cell_type', bulk_group=None, top_marker_num=200, ratio_num=1,gpu=0)
```

```python
CellFractionPrediction=model.predicted_fraction()
```

```python
ax = CellFractionPrediction.plot(kind='barh', stacked=True, figsize=(8, 4),)
ax.set_xlabel('Sample')
ax.set_ylabel('Cell Fraction')
ax.set_title('TAPE Cell fraction predicted')
plt.legend(bbox_to_anchor=(1.05, 1),ncol=1,)
plt.show()
```

```python
model.bulk_preprocess_lazy()
model.single_preprocess_lazy()
model.prepare_input()
```

```python
vae_net=model.train(
    batch_size=512,
    learning_rate=1e-4,
    hidden_size=256,
    epoch_num=3500,
    vae_save_dir="D:scanpy/model",
    vae_save_name='Capsaicn_June',
    generate_save_dir="D:scanpy/model/output",
    generate_save_name='Capsaicin_June_28')
```

```python
model.plot_loss()
```
